```python
import math
def encode(message):
	encoded_message = ""
	i = 0

	unique_characters = set(message)
	bits_required = math.ceil(math.log(len(unique_characters), 2))

	j = 0
	for val in unique_characters:
		print("code for {} = {}".format(val,bin(j).replace("0b", "").rjust(bits_required,'0')))
		j = j+1

	while (i <= len(message)-1):
		count = 1
		ch = message[i]
		j = 0
		for val in unique_characters:  
			if (ch == val):
				encoded_message=encoded_message+ bin(j).replace("0b", "").rjust(bits_required,'0')
				break
			else:
				j = j+1
		i = i+1
	return encoded_message, bits_required, unique_characters

#open text file in read mode
text_file = open("codingDataExample.txt", "r")
 
#read whole file to a string
message = text_file.read()
 
#close file
text_file.close()

encoded_message, bits_required, unique_characters = encode(message)
print("Number of bits used per character for fixed length coding = {}".format(bits_required))
print("Assuming that 8 bits are used to encode evey character in the uncompressed file.")
print("Memory requirement for uncompressed message : {} bits".format(len(message)*8))
print("Memory requirement for compressed message : {} bits".format(len(encoded_message)))
print("Cmpression ratio = {:.5}".format(len(message)*8/len(encoded_message)))
print("Encoded message = {}".format(encoded_message))
```
