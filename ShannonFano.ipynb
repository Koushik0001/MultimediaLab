```python
import json
import csv
```

```python

def buildFirstNode(message):
    charSet = {}
    totalCharacterCount = 0
    
    for char in message:
        totalCharacterCount+=1
        if char in charSet:
            charSet[char] += 1
        else:
            charSet[char] = 1
    
    for char in charSet:
        charDataObj = {}
        charDataObj["symbol"] = char
        charDataObj["count"] = charSet[char]
        charDataObj["frequency"] = charSet[char]/totalCharacterCount 
        tree["node"].append(charDataObj)
    
    tree["node"].sort(key=lambda char : char["frequency"], reverse=True)
```

```python
def findDivisionPoint(arr):#array of nodes is passed
    n = len(arr)
    left_sum = arr[0]["frequency"]
    right_sum = sum(d["frequency"] for d in arr[1:])
    min_diff = abs(left_sum - right_sum)
    min_index = 0
    for i in range(1, n):
        left_sum += arr[i]["frequency"]
        right_sum -= arr[i]["frequency"]
        diff = abs(left_sum - right_sum)
        if diff < min_diff:
            min_diff = diff
            min_index = i
    return arr[:min_index+1], arr[min_index+1:]
 
```

```python
def recurssiveFun(node, children, code):
    subarray1, subarray2 = findDivisionPoint(node)
    children.append({
           "node" : subarray1,
           "children" : [],
           "code" : code+"0"
        })
    children.append({
           "node" : subarray2,
           "children" : [],
           "code" : code+"1"
        })
    if len(children[0]["node"])>1:
        recurssiveFun(children[0]["node"],children[0]["children"], children[0]["code"])
    else:
        print()
        charDataObj = {"count":children[0]["node"][0]["count"], "frequency": children[0]["node"][0]["frequency"],"code": children[0]["code"]}
        codes[children[0]["node"][0]["symbol"]] = charDataObj

    if len(children[1]["node"])>1:
        recurssiveFun(children[1]["node"],children[1]["children"], children[1]["code"])
    else:
        charDataObj = {"count":children[1]["node"][0]["count"], "frequency": children[1]["node"][0]["frequency"], "code": children[1]["code"]}
        codes[children[1]["node"][0]["symbol"]] = charDataObj
```

```python
def encode(message):
    buildFirstNode(message)
    recurssiveFun(tree["node"], tree["children"], tree["code"])
```
